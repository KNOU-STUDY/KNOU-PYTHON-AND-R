```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,6,0.1)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,6,0.1)
```

```python
y1 = np.sin(x)
```

```python
y2 = np.cos(x)
```

```python
plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle='--',label='cos')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('sin & cos')
plt.legend()
plt.show()
```
